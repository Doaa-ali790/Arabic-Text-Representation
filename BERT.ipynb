```python
!pip install transformers torch

import torch
from transformers import AutoTokenizer, AutoModel
import numpy as np

# سورة الفاتحة
verses = [
    "بسم الله الرحمن الرحيم",
    "الحمد لله رب العالمين",
    "الرحمن الرحيم",
    "مالك يوم الدين",
    "إياك نعبد وإياك نستعين",
    "اهدنا الصراط المستقيم",
    "صراط الذين أنعمت عليهم غير المغضوب عليهم ولا الضالين"
]

# تحميل BERT العربي الحقيقي
tokenizer = AutoTokenizer.from_pretrained("asafaya/bert-base-arabic")
model = AutoModel.from_pretrained("asafaya/bert-base-arabic")

results = []

# استخراج embeddings
for i, verse in enumerate(verses, 1):
    inputs = tokenizer(verse, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)

    embeddings = outputs.last_hidden_state.squeeze(0).numpy()
    results.append((i, embeddings))

# حفظ النتائج في ملف نصي
with open("bert_results.txt", "w", encoding="utf-8") as f:
    for verse_num, emb in results:
        f.write(f"Verse {verse_num} Embeddings (shape={emb.shape}):\n")
        f.write(str(emb))
        f.write("\n\n")

print("تم إنشاء ملف bert_results.txt بنجاح")
```
